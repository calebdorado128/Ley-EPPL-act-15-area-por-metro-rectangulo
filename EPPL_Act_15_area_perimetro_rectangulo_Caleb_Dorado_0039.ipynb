1. Redacción del Problema: Análisis GeométricoEl rectángulo es un paralelogramo cuyos cuatro lados forman ángulos rectos entre sí ($90^\circ$).
* En el ámbito de la geometría analítica y el desarrollo de software educativo, es fundamental automatizar el cálculo de sus propiedades métricas para evitar errores humanos y optimizar el aprendizaje.
* Definición de las VariablesPara que el sistema funcione correctamente, el usuario debe proveer dos dimensiones esenciales:Base ($b$): La medida horizontal del lado inferior.Altura ($h$): La medida vertical perpendicular a la base.
* Fórmulas a utilizarEl software implementará las siguientes funciones matemáticas para procesar los datos de entrada:Área ($A$): Representa la superficie encerrada dentro de los límites del rectángulo.
* Se mide en unidades cuadradas (como $cm^2$ o $m^2$).$$A = b \times h$$Perímetro ($P$): Es la longitud total del contorno de la figura. Se obtiene sumando la longitud de sus cuatro lados.
* Dado que los lados opuestos son iguales, la fórmula se simplifica a:$$P = 2b + 2h = 2 \cdot (b + h)$$

```python
# ==========================================
# programa echo por Caleb Dorado 0039
# FUNCIONES DE CÁLCULO
# ==========================================

def calcular_area_rectangulo(base, altura):
    """Calcula y muestra el área."""
    area = base * altura
    print("\n" + "="*30)
    print(f" RESULTADO DEL ÁREA")
    print(f" Fórmula: {base} * {altura}")
    print(f" Total: {area} unidades cuadradas")
    print("="*30)

def calcular_perimetro_rectangulo(base, altura):
    """Calcula y muestra el perímetro."""
    perimetro = 2 * (base + altura)
    print("\n" + "="*30)
    print(f" RESULTADO DEL PERÍMETRO")
    print(f" Fórmula: 2 * ({base} + {altura})")
    print(f" Total: {perimetro} unidades lineales")
    print("="*30)

# ==========================================
# FUNCIÓN PRINCIPAL Y MENÚ
# ==========================================

def mostrar_menu():
    """Maneja la lógica de la interfaz de usuario."""
    ejecutando = True

    while ejecutando:
        print("\n--- CALCULADORA GEOMÉTRICA ---")
        print("1. Calcular Área")
        print("2. Calcular Perímetro")
        print("3. Salir")

        opcion = input("\nSeleccione una opción (1, 2 o 3): ")

        # Lógica de salida
        if opcion == '3':
            print("\nFinalizando programa. ¡Que tengas un excelente día!")
            ejecutando = False

        # Lógica de procesamiento
        elif opcion == '1' or opcion == '2':
            try:
                # Solicitud de datos al usuario
                b = float(input("Ingrese el valor de la base: "))
                h = float(input("Ingrese el valor de la altura: "))

                if b <= 0 or h <= 0:
                    print("\n⚠️ Error: Las medidas deben ser mayores a cero.")
                elif opcion == '1':
                    calcular_area_rectangulo(b, h)
                elif opcion == '2':
                    calcular_perimetro_rectangulo(b, h)

            except ValueError:
                print("\n⚠️ Error: Por favor, ingrese solo valores numéricos.")

        # Opción no contemplada
        else:
            print("\n⚠️ Opción no válida. Intente con los números 1, 2 o 3.")

# Iniciar la aplicación
if __name__ == "__main__":
    mostrar_menu()
```
